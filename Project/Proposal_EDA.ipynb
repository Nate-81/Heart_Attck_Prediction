# Superficial EDA

## Correlation Matrix of the Numerical Columns

This is just the correlation matrix of the numerical columns, The main project will contain a correlation matrix for all the used columns, since that would take a lot of work to Ordinal/One-Hot Encode right now.

```python
df_numerical = df.select_dtypes(np.number)
```

```python
corr = df_numerical.corr()
```

```python
cmap = sns.color_palette("viridis", as_cmap=True)
fig,ax = plt.subplots(figsize=(20,15))
ax = sns.heatmap(corr,annot=True,cmap=cmap)
ax.set_title("Correlation Matrix of the Numerical Columns\n",fontsize=25)
plt.show()
```

It is nice to know that, there is no highly correlated data within this dataset that needs to be removed. The only somewhat correlated data is between smoking and age, which makes sense.

## Scanning for Outliers

Since there is hardly any correlation in the data, scatter plots wont, be as useful. However,the EDA that we can still do is look through come of the columns and check them for outliers

```python
fig,axes = plt.subplots(2,2,figsize=(10,10))
sns.boxplot(ax = axes[0,0], data=df["Sleep Hours Per Day"])
axes[0,0].set_title("Sleep Hours Per Day")
sns.boxplot(ax = axes[0,1], data=df["Income"])
axes[0,1].set_title("Income")
sns.boxplot(ax = axes[1,0], data=df["Physical Activity Days Per Week"])
axes[1,0].set_title("Physical Activity Days Per Week")
sns.boxplot(ax = axes[1,1], data=df["Cholesterol"])
axes[1,1].set_title("Cholesterol")
```

It seems somehow that, the data had already been cleaned, and the outliers taken care of, for these four columns. The project will go into further detail on outliers for other columns, to definitely make sure there are none.
